```python
df_2010.drop(['Rank'], inplace=True, axis=1)
#df_2010.set_index('Language', inplace=True)
df_2010.rename(columns={'Share':'2010'}, inplace=True)
df_2010.head()
```

```python
df_2011.drop(['Rank'], inplace=True, axis=1)
#df_2011.set_index('Language', inplace=True)
df_2011.rename(columns={'Share':'2011'}, inplace=True)
df_2011.head()
```

```python
df_2012.drop(['Rank'], inplace=True, axis=1)
#df_2012.set_index('Language', inplace=True)
df_2012.rename(columns={'Share':'2012'}, inplace=True)
df_2012.head()
```

```python
df_2013.drop(['Rank'], inplace=True, axis=1)
#df_2013.set_index('Language', inplace=True)
df_2013.rename(columns={'Share':'2013'}, inplace=True)
df_2013.head()
```

```python
df_2014.drop(['Rank'], inplace=True, axis=1)
#df_2014.set_index('Language', inplace=True)
df_2014.rename(columns={'Share':'2014'}, inplace=True)
df_2014.head()
```

```python
df_2015.drop(['Rank'], inplace=True, axis=1)
#df_2015.set_index('Language', inplace=True)
df_2015.rename(columns={'Share':'2015'}, inplace=True)
df_2015.head()
```

```python
df_2016.drop(['Rank'], inplace=True, axis=1)
#df_2016.set_index('Language', inplace=True)
df_2016.rename(columns={'Share':'2016'}, inplace=True)
df_2016.head()
```

```python
df_2017.drop(['Rank', 'Trend'], inplace=True, axis=1)
#df_2017.set_index('Language', inplace=True)
df_2017.rename(columns={'Share':'2017'}, inplace=True)
df_2017.head()
```

```python
df_2018.drop(['Rank', 'Trend'], inplace=True, axis=1)
#df_2018.set_index('Language', inplace=True)
df_2018.rename(columns={'Share':'2018'}, inplace=True)
df_2018.head()
```

```python
df_2019.drop(['Rank', 'Trend'], inplace=True, axis=1)
#df_2019.set_index('Language', inplace=True)
df_2019.rename(columns={'Share':'2019'}, inplace=True)
df_2019.head()
```

```python
df_2010['2010']=df_2010['2010'].replace('\%','',regex=True).astype(float)
df_2011['2011']=df_2011['2011'].replace('\%','',regex=True).astype(float)
df_2012['2012']=df_2012['2012'].replace('\%','',regex=True).astype(float)
df_2013['2013']=df_2013['2013'].replace('\%','',regex=True).astype(float)
df_2014['2014']=df_2014['2014'].replace('\%','',regex=True).astype(float)
df_2015['2015']=df_2015['2015'].replace('\%','',regex=True).astype(float)
df_2016['2016']=df_2016['2016'].replace('\%','',regex=True).astype(float)
df_2017['2017']=df_2017['2017'].replace('\%','',regex=True).astype(float)
df_2018['2018']=df_2018['2018'].replace('\%','',regex=True).astype(float)
df_2019['2019']=df_2019['2019'].replace('\%','',regex=True).astype(float)
```

```python
df_2010['2010']=df_2010['2010']*100/df_2010['2010'].sum()
df_2011['2011']=df_2011['2011']*100/df_2011['2011'].sum()
df_2012['2012']=df_2012['2012']*100/df_2012['2012'].sum()
df_2013['2013']=df_2013['2013']*100/df_2013['2013'].sum()
df_2014['2014']=df_2014['2014']*100/df_2014['2014'].sum()
df_2015['2015']=df_2015['2015']*100/df_2015['2015'].sum()
df_2016['2016']=df_2016['2016']*100/df_2016['2016'].sum()
df_2017['2017']=df_2017['2017']*100/df_2017['2017'].sum()
df_2018['2018']=df_2018['2018']*100/df_2018['2018'].sum()
df_2019['2019']=df_2019['2019']*100/df_2019['2019'].sum()
```

```python
df=pd.merge(df_2010, df_2011, on='Language', how='outer')
df=pd.merge(df, df_2012, on='Language', how='outer')
df=pd.merge(df, df_2013, on='Language', how='outer')
df=pd.merge(df, df_2014, on='Language', how='outer')
df=pd.merge(df, df_2015, on='Language', how='outer')
df=pd.merge(df, df_2016, on='Language', how='outer')
df=pd.merge(df, df_2017, on='Language', how='outer')
df=pd.merge(df, df_2018, on='Language', how='outer')
df=pd.merge(df, df_2019, on='Language', how='outer')
df
```

```python
X = df[['2010','2011','2012','2013','2014','2015','2015','2016','2017','2018']].values
y = df['2019'].values
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=0)
regressor = lm.LinearRegression()  
regressor.fit(X_train, y_train) #training the algorithm
#To retrieve the intercept:
print(regressor.intercept_)
#For retrieving the slope:
print(regressor.coef_)
```

```python
y_pred = regressor.predict(X_test)
df_1 = pd.DataFrame({'Actual': y_test.flatten(), 'Predicted': y_pred.flatten()})
df_1
```

```python
df1 = df_1.head()
df1.plot(kind='bar',figsize=(16,10))
plt.grid(which='major', linestyle='-', linewidth='0.5', color='green')
plt.grid(which='minor', linestyle=':', linewidth='0.5', color='black')
plt.show()
```
